## Helper Functions

```python
# Wrapper/Helper Functions
def biplot(data, plot_type, x, y, figure_size, title, palette_1, palette_2, hue_1, hue_2):
    '''
    Wrapper Function to create a bi-plot. 
    
    Arguments: 
        data: a Pandas dataframe 
        plot_type: specify the type of plot as scattter, box, violin, histogram, or bar
        x: the x-axis from the Pandas dataframe 
        y: the y-axis from the Pandas dataframe, if applicable 
        figure_size: a tuple representing the dimensions of the plot 
        title: the title of the plot 
        palette_1 and palette_2: the color paletter scheme 
        hue_1 and hue_2: the hue, used to color the data by a category 
    '''
    f = plt.figure(figsize = figure_size)
    ax = f.add_subplot(121)
    if plot_type == 'scatter':
        sns.scatterplot(data = data, x = x, y = y, palette = palette_1, hue = hue_1, ax = ax)
    elif plot_type == 'boxplot':
        sns.boxplot(data = data, x = x, y = y, palette = palette_1, hue = hue_1, ax = ax)
    ax.set_title(title, fontsize = 16)
        
    ax = f.add_subplot(122) 
    if plot_type == 'scatter':
        sns.scatterplot(data = data, x = x, y = y,  palette = palette_2, hue = hue_2)
    elif plot_type == 'boxplot':
        sns.boxplot(data = data, x = x, y = y, palette = palette_2, hue = hue_2, ax = ax)
    ax.set_title(title, fontsize = 16)  
    #plt.savefig('sc.png') 
```

__Outlier Detection by Cooke's Distance__

```python
import statsmodels.api as sm
from sklearn.preprocessing import LabelBinarizer
```

First, I need to one-hot encode categorical variables and drop any NaNs.  

```python
# convert categorical variables to dummy variables
df_b1 = df_b.copy()
label_binarizer = LabelBinarizer() 

categorical_variables = ['sex', 'smoker', 'region']
label_binarizer_outputs = []

# for loop to binarize each 
for category in categorical_variables:
    label_binarizer_outputs.append(
        label_binarizer.fit_transform(
            df_b1[category]
        ) 
    )
```

```python
# Create a new dataframe that includes one-hot encoded variable
df_b2 = df_b1[['charges', 'age', 'bmi', 'children']]
columns = [ ['sex'], ['smoker'], ['region1', 'region2', 'region3', 'region4']]
i = 0
for output in label_binarizer_outputs: 
    # if we have a matrix of one-hot encoded variables...
    if len(columns[i]) > 1: 
        # ... first proceed as normal.  Copy and convert the matrix to a dataframe.  
        tmp_df = pd.DataFrame(data = output, columns = columns[i])
        # for each of the encoded categories, append to the copied dataframe (df_b2)
        k = 0
        for region in columns[i]:
            df_b2[ columns[i][k] ] = tmp_df[ columns[i][k] ]
            k += 1
    else:
        tmp_df = pd.DataFrame(data = output, columns = columns[i])
        df_b2[ columns[i][0] ] = tmp_df[ columns[i][0] ]
        i += 1
```

Now deal with any NaNs. 

```python
# It turns out that we end up with a bunch of NaNs, so we need to remove them here. 
df_b3 = df_b2.dropna()
```

Fit a model that I can use to evaluate observations on the basis of Cooke's Distance.

```python
# Fit a model that we will use to evaluate observations on the basis of Cooke's Distance.
y = df_b3['charges']
X = df_b3[['age', 'bmi', 'children', 'sex', 'smoker', 'region1', 'region2', 'region3', 'region4']]
X = sm.add_constant(np.asarray(X))

model = sm.OLS(y, X).fit()
```

```python
np.set_printoptions(suppress = True)
influence = model.get_influence()
cooks = influence.cooks_distance
```

```python
plt.scatter(df_b3['charges'], cooks[0])
plt.xlabel('Charges')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
df_b3['Cooks_Distance'] = pd.DataFrame(cooks[0])
```

```python
df_b3.head()
```

```python
cook_threshold = 4/548

# filter outliers
df_no = df_b3[(df_b3['Cooks_Distance'] < cook_threshold)]
```

```python
df_no['charges'].describe()
```

```python
sns.boxplot(data = df_no['charges']).set(title = "Boxplot of Medicare Charges, Filtered for Outliers")
```

```python
f= plt.figure(figsize=(12,4))

sns.histplot(df_no['charges'],bins = 50, color= 'r').set(title = "Distribution of Medicare Charges, Filtered for Outliers")
```

It's not perfect, but these observations do appear to range a bit more consistently. The biggest concern now is that we may have too few observations to accurately simulate real-world conditions.  At 70 observations, I feel like I'm cutting it a bit close, but I think it might be just enough to work with.  I think what I might do is try to build two models: one model built with a dataset that hasn't been pared down to exclude outliers, and one built on a training set that has everything.

## 1st Generation Model 

Built on the dataset that was pared down using Cooke's Distance.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Separate dataset out into an X and y set
X = df_no['age']
y = df_no['charges']

# Convert to a numpy array
X = np.array(X)
y = np.array(y)

# Reshape into a compatible format
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

# Split into Training and Testing Datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Intercept and Coefficient 
print("The intercept for the model is " + str(lin_reg.intercept_) + " and the coefficient is " + str(lin_reg.coef_) +".")
```

#### Checking for Homoscedasticity 
Residual Plot to look for the presence of Heteroscedasticity, which occurs when the residuals are not identically distributed (constant).

```python
# Get predictions and calculate residuals
y_pred = lin_reg.predict(X_test)
residuals = y_test - y_pred
```

```python
# Plot Residuals
plt.scatter(y_pred, residuals)
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residuals vs. Fitted Values Plot')

plt.show()
```

These residuals do not look identically distributed, so it is likely our model fails this test.

#### Checking Normality Assumptions 
Visualize Residuals vs Quantiles to assess normality of residuals.  

```python
fig = sm.qqplot(residuals, line='45')
plt.title('Normal QQ-Plot')

plt.show()
```

The residuals are not normally distributed.

#### Evaluating the Model 
Calculate the R-Squared and Mean Squared Error to evaluate goodness of fit and performance on test dataset

```python
from sklearn.metrics import mean_squared_error
```

```python
print("R-Squared: " + str(round(lin_reg.score(X_train, y_train), 4)))
print("Mean Squared Error: " + str( round( mean_squared_error(y_test, y_pred), 4)))
```

Overall, this model performs quite poorly.  The R-Squared is quite low, and although we do not have other models to compare against, the Mean Squared Error on the generation one model is quite high.  

If this was a real-world scenario, I would not reccomend using this in practice.

## Conclusions

It is proving difficult to build a reliable regressor model on this data.  Many of the expectations that are required by well-functioning linear regression models, including linearity, homoscedasticity, and normality of residuals, are not met, or are weakly met, as is the case with the linearity expectation.  There is some scant evidence to suggest that perhaps a non-linear regressor model, such as a Random Forest Regressor, could serve as the basis of a better regressor model.  Regardless, I think building a successful regressor model off of this data would be extremely difficult.   

The data itself is widely inconsistent, with massive outliers, and the correlations between independent and dependent variables are weak, even when they are visible.  Age does appear to have a very recognizable linear relationship with Charges, but it is separated into three distinct lines, which could suggest that there is a missing categorical variable.  Smoking Status appears to have a clear relationship in exploratory analysis, with Smokers appearing to have a higher baseline Charge rate than Non-Smokers.  BMI has a weaker relationship than Age with Charge Rates, and appears to fork in two different directions, which I think could suggest some sort of missing categorical variable.  And finally, there is some evidence to suggest that the regions Northwest, Northeast, and Southwest tend to have lower average Charges than Southeast.  Aside from these variables, nothing else appears to have any meaningful influence on the Charge rate.  

In conclusion, while there is evidence to suggest BMI, Age, Smoking Status, and perhaps Region might influence Charge rates, there is not enough evidence to make that a foregone conclusion.  Part of the problem is the dataset itself; it could be missing important categorical variable that could be used to build a strong model that demonstrates a more direct causal link.  The dataset also has inconsistencies that make it hard to build well functioning models, including widely varying charge rates with seemingly no consistent distribution, and large imbalances between the number of smokers and non-smokers.
